```python
import numpy as np          # Numpy es para arreglos numéricos
import pylab as plt         # Pylab es para graficación 
import pandas as pd         # Pandas es para manejo de Tablas de datos
```

```python
data = pd.read_csv('/content/ibtracs.NA.list.v04r00.csv', sep=',',dtype={'NAME': str, 'USA_STATUS':str})
```

```python
data
```

```python
best_track = data.iloc[:, [5,6,8,9,10,11,22]] 
best_track.drop(0, axis=0, inplace=True)
best_track
```

```python
vel =  pd.to_numeric(data['WMO_WIND'],errors='coerce')
vel_kmh= vel*1.852
vel_kmh 
```

```python
best_track['ISO_TIME']
```

```python
data_K = pd.to_datetime(best_track['ISO_TIME'])


dates = pd.DatetimeIndex(data_K)
dates

años = dates.year
meses = dates.month
dias = dates.day
horas = dates.hour

evento = data['NAME']


años = pd.to_numeric(años,downcast='integer',errors='coerce')
meses = pd.to_numeric(meses,errors='coerce')
dias = pd.to_numeric(dias,errors='coerce')
horas = pd.to_numeric(horas,errors='coerce')
pres = pd.to_numeric(data['WMO_PRES'],errors='coerce')
lat = pd.to_numeric(data['LAT'],errors='coerce')
lon = pd.to_numeric(data['LON'],errors='coerce')
typemax = pd.to_numeric(data['USA_SSHS'],errors='coerce',downcast='integer')
```

```python
fin = [evento,dias, meses, años, lat, lon,horas, vel_kmh, pres,typemax ]
fin = pd.DataFrame(fin)

fin = fin.T
fin

df_new = fin.rename(columns={'NAME': 'event','LAT':'lat','LON':'lon','WMO_WIND':'windkph','WMO_PRES':'press','USA_SSHS':'typemax' })
print(df_new)
```

```python
df_new.to_csv(r'c:\data\huralt.txt', sep='\t', mode='a')
```
